## Step 4: K-Means Silhouette Score

```python
silhouette_scores = []
valid_k = []
for k in k_values:
    if k == 1:
        continue  # Silhouette score is not defined for k=1
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    valid_k.append(k)

plt.figure(figsize=(8, 5))
plt.plot(valid_k, silhouette_scores, marker='o')
plt.title('Silhouette Score for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.show()
```

 ## Step 5: K-Means Clustering

```python
kmeans = KMeans(n_clusters=23, random_state=42)
kmeans.fit(X)

# Attach cluster labels back to product names
product_clusters = pd.DataFrame({
    'product_name': product_names,
    'cluster': kmeans.labels_
})
# Step 4: View grouped products by cluster
for i in range(kmeans.n_clusters):
    print(f"\nCluster {i}:\n")
    print(product_clusters[product_clusters['cluster'] == i]['product_name'].tolist())
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
reduced = pca.fit_transform(X.toarray())

# Plot
plt.figure(figsize=(12, 8))
plt.scatter(reduced[:, 0], reduced[:, 1], c=kmeans.labels_, cmap='tab10', s=50)
for i, name in enumerate(product_names):
    plt.text(reduced[i, 0], reduced[i, 1], name[:15], fontsize=8)
plt.title('Product Clustering (PCA 2D View)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.colorbar(label='Cluster ID')
plt.grid(True)
plt.show()
```


### ðŸ” Updated K-Means Clustering with Categorization

This section extends the existing TF-IDF-based clustering by:
- Cleaning product names
- Reusing improved product categories
- Connecting cluster labels to products
- Showing cluster distributions by improved category


```python

from sklearn.feature_extraction.text import TfidfVectorizer

product_names = df['name'].fillna('').unique()
product_names = [name.lower() for name in product_names]  # Lowercase cleaning

vectorizer = TfidfVectorizer(
    max_features=1000,
    stop_words='english',
    ngram_range=(1, 2)
)
X = vectorizer.fit_transform(product_names)

print(f"TF-IDF Matrix Shape: {X.shape}")
```

```python

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

Cluster_labels = []
k_values = range(2, 23)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    Cluster_labels.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_values, Cluster_labels, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Cluster_labels')
plt.show()
```

```python

final_k = 8  # Use the elbow method result here
kmeans = KMeans(n_clusters=final_k, random_state=42)
labels = kmeans.fit_predict(X)

# Map cluster labels back to products
name_to_label = dict(zip(product_names, labels))
df['cluster_label'] = df['name'].map(name_to_label)
```

```python

summary = df.groupby(['meta_category_improved', 'cluster_label']).size().unstack(fill_value=0)
print("Cluster distribution by improved category:")
print(summary)
```
